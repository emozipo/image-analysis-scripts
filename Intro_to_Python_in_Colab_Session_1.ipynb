# For loops

Python has a number of ways to iterate, or "loop" over commands. The most common is the `for` loop, which allows a single block of code to be executed multiple times in sequence. We can use a `for` loop to repeat a chunk of code for every element in a list like so:


````python
strings = ["Hello!", "How", "are", "you", "doing?"]

# our first for loop!
for item in strings:
  print(item)

print("All done!")
````

Note the special syntax used in the for loop. For loops must have:

- a first line with structure
`for (loop variable) in (collection):`
- a colon at the end of the first line
- an indented block of code below it to repeat in the loop

In the loop above, for each element in the list `strings`, assign the element to the loop variable `item` and then execute the code indented under the first line. This means that the statement `print(item)` is executed 5 times, each time with a different value assigned to `item`.

## If and else

Suppose we wanted to execute some code if a condition is met and execute some other code if that condition is *not* met.

It's easy to do this with boolean operators:


````python
my_string = "Hello world!"

if 'a' in my_string:
    print("Your string has an 'a' in it!")
if not 'a' in my_string:
    print("Your string doesn't have an 'a' in it...")
````

However, it's a bit cumbersome to write out the exact same condition just to negate it. Fortunately, Python provides a simple alternative with the `else` keyword! Simply add an `else` branch at the end of any `if` statement, like so:



````python
my_string = "Hello world!"

if 'a' in my_string:
    print("Your string has an 'a' in it!")
else:
    print("Your string doesn't have an 'a' in it...")
````

You can also check multiple different conditions using `if`. The simplest way is to use multiple `if` blocks, which checks each condition independently.



````python
user_type = "admin user"

if "admin" in user_type:
    print("You are an admin! Wow!")
if "user" in user_type:
    print("You are a user.")
````

But what if we want to only print *once* based on the highest clearance level of the user? In that case, we could do something like this:

````python
user_type = "admin user"

if "admin" in user_type:
    print("You are an admin! Wow!")
else:
    if "user" in user_type:
        print("You are a user.")
````

This pattern of "If X is true, do this. Otherwise, if Y is true, do that." is common enough that Python provides an easy way to do it using the `elif` keyword (short for "else if"). This next code cell does the exact same thing as the previous one:

````python
user_type = "admin user"

if "admin" in user_type:
    print("You are an admin! Wow!")
elif "user" in user_type:
        print("You are a user.")
````

You can chain as many `elif`s as you want in an `if` block. You can also add an `else` to the end if you'd like.

````python
user_type = "power user"

if "admin" in user_type:
    print("You are an admin! Wow!")
elif "power" in user_type:
    print("You are a power user! Cool.")
elif "user" in user_type:
    print("You are a user.")
else:
    print("You are logged out.")
````

## Understanding nested expressions

Function calls and arithmetic expressions can themselves contain expressions.

Nested expressions can turn into complicated-looking code. However, the way in which complicated expressions break down is very regular.

Suppose we are interested in heights that are very unusual. We'll say that a height is unusual to the extent that it's far away on the number line from the average human height. [An estimate](http://press.endocrine.org/doi/full/10.1210/jcem.86.9.7875?ck=nck&) of the average adult human height is 1.688 meters.

So if Kayla is 1.21 meters tall, then her height is $\lvert 1.21 - 1.688 \rvert$, or $0.478$, meters away from the average. Here's a picture of that:


image.png

And here's how we'd write that in one line of Python code:


````python
abs(1.21 - 1.688)
````

What's going on here? `abs` takes just one argument, so the stuff inside the parentheses is all part of that *single* argument. Specifically, the argument is the value of the expression `1.21 - 1.688`. The value of that expression is `-.478`. That value is the argument to `abs`. The absolute value of that is `.478`, so `.478` is the value of the full expression `abs(1.21 - 1.688)`.

Picture simplifying the expression in several steps:

1. `abs(1.21 - 1.688)`
2. `abs(-.478)`
3. `.478`

In fact, that's basically what Python does to compute the value of the expression.

### More nesting

Now say that we want to compute the more unusual of the two heights. We'll use the function `max`, which (as you may recall) takes two numbers as arguments and returns the larger of the two arguments. Combining that with the `abs` function, we can compute the larger distance from average among the two heights.

In the following example, suppose we have another person, Paola, who is 1.76m tall.



````python
# Just read and run this cell.

kayla_height_m = 1.21
paola_height_m = 1.76
average_adult_height_m = 1.688

# The larger distance from the average human height, among the two heights.
larger_distance_m = max(abs(kayla_height_m - average_adult_height_m), abs(paola_height_m - average_adult_height_m))

# Print out our results in a nice readable format. Pay close attention to how
# the `print` function behaves when given multiple arguments.
print("The larger distance from the average height among these two people is", larger_distance_m, "meters.")
````

The line where `larger_distance_m` is computed looks complicated, but we can break it down into simpler components just like we did before.

The basic recipe is to repeatedly simplify small parts of the expression:

- **Basic expressions**: Start with expressions whose values we know, like variables or numbers.
  - Examples: `paola_height_m` or `5`.
- **Find the next simplest group of expressions**: Look for basic expressions that are directly connected to each other. This can be by arithmetic or as arguments to a function call.
  - Example: `kayla_height_m - average_adult_height_m`.
- **Evaluate that group**: Evaluate the arithmetic expression or function call. Use the value computed to replace the group of expressions.
  - Example: `kayla_height_m - average_adult_height_m` becomes `-.478`.

This code extracts the substring `'This'` from `my_string` by specificing the start index `0`. The character at `my_string[0]` will be included in the string. For the stop index, we specify the index *immediately after* the last character that we want in the string: `4`.

The distinction between inclusive and exclusive indices may seem arbitrary, but it means that the difference between the start and stop indices is always equal to the length of the string we want to extract!

## Boolean operators

Often, we'd like to check more than a single condition within an `if` statement. For example, to check whether a variable `x` is between two numbers `a` and `b`, we would need to check two conditions: `x >= a` and `x <= b`.

Python has three *boolean operators* that we can use to manipulate booleans.

The simplest operator, `not`, takes a single operand and "flips" it:

````python
not True
````

````python
x = 5
not x < 5
````

The `and` operator takes two operands and returns `True` only if *both* its operands are `True`. Otherwise, it returns `False`.


````python
True and False
````

````python
# The `in` keyword checks if a value is included in some sequence
'H' in 'Hello' and 'o' in 'Hello'
````

Finally, the `or` operator takes two operands and returns `True` if *either* of its operands are `True`. It only returns `False` if both its operands are `False`.

````python
True or False
````

````python
x = 5
x > 5 or x < 5
````
